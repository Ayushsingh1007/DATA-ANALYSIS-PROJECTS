### Handling Missing Values in **DATA**

```python
data.dropna(subset=['name', 'host_name'], inplace=True)

data['reviews_per_month'].fillna(0, inplace=True)

if 'last_review' in data.columns:
    data.drop('last_review', axis=1, inplace=True)

print("Missing values after cleaning:")
print(data.isna().sum())
```

Now that we've addressed the missing values, let's take a look at the updated information and first few rows of the cleaned **data** DataFrame.

```python
data.info()
display(data.head())
```

**PLOTTING SCATTER MATRIX

```python
pd.plotting.scatter_matrix(data, figsize=(20, 10))
plt.show()
```

**Correlation matrix:**

```python
correlation_matrix = data.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Data')
plt.show()
```

**Distribution graphs (histogram/bar graph) of sampled columns:**

```python
num_cols = data.select_dtypes(include=np.number).columns
fig, axes = plt.subplots(nrows=len(num_cols) // 5 + 1, ncols=5, figsize=(20, 4 * (len(num_cols) // 5 + 1)))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.histplot(data[col], kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(data['price'], data['number_of_reviews'], data['availability_365'])
ax.set_xlabel('Price')
ax.set_ylabel('Number of Reviews')
ax.set_zlabel('Availability 365')
ax.set_title('3D Scatter Plot of Price vs. Reviews vs. Availability')

plt.show()
```
